```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits,on="title")
```

```python
movies
```

```python
movies=movies[['genres','movie_id','keywords','title','overview','popularity','runtime','vote_average','vote_count','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.duplicated().sum()
```

```python
def seperate(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
import ast
```

```python
movies['genres']=movies['genres'].apply(seperate)
movies['keywords']=movies['keywords'].apply(seperate)
```

```python
def getCharacters(data):
    l=[]
    counter=0
    for i in ast.literal_eval(data):
        if counter!=5:
            l.append(i['name'])
            counter+=1;
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(getCharacters)
```

```python
def getDirector(data):
    l=[]
    for i in ast.literal_eval(data):
        if i['job']=='Director':
            l.append(i['name'])
    return l
```

```python
movies['crew']=movies['crew'].apply(getDirector)
import pickle
pickle.dump(movies.to_dict(),open('movies_original_dict.pkl','wb'))
movies
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies['tags']=movies['tags'].apply(lambda x:" ".join(x))
movies['tags']=movies['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(data):
    y=[]
    for i in data.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
movies
```

```python
#movies['crew']=movies['crew'].apply(lambda x:" ".join(x))
```

```python
C=movies['vote_average'].mean()
```

```python
m=movies['vote_count'].quantile(0.9)
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
movies['score']=movies.apply(weighted_rating,axis=1)
```

```python
movies
```

```python
movies_list=movies.sort_values('score',ascending=False)
director_list=list(enumerate(movies_list['crew']))
director_list
for i in director_list:
    if 'James Cameron' in i[1]:
        print(movies.iloc[i[0]])
```

```python
def recommendUsingDirectors(Director):
    Director=Director.replace(" ","")
    movie_list=movies.sort_values('score',ascending=False)
    director_list=list(enumerate(movie_list['crew']))
    
    for i in director_list:
        if Director in i[1]:
            print(movie_list.iloc[i[0]].title)
```

```python
recommendUsingDirectors('James Cameron')
```

```python
tfidf_matrix = tfidf.fit_transform(movies['tags'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distances=cosine_sim[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    
    for i in movie_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
def recommendUsingCast(character):
    character=character.replace(" ","")
    movie_list=movies.sort_values('score',ascending=False)
    actors_list=list(enumerate(movie_list['cast']))

    for i in actors_list:
        if character in i[1]:
            print(movie_list.iloc[i[0]].title)
```

```python
recommendUsingCast('Sam Worthington')
```

```python
def recommendUsingGenre(genre):
    genre=genre.replace(" ","")
    movie_list=movies.sort_values('score',ascending=False)
    genres_list=list(enumerate(movie_list['genres']))
    
    counter=0

    for i in genres_list:
        if genre in i[1]:
            if counter!=10:
                print(movie_list.iloc[i[0]].title)
                counter+=1
```

```python
recommendUsingGenre('Drama')
```
